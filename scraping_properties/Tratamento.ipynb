```python
import pandas as pd
import numpy as np
import re

df = pd.read_excel('casinhas.xlsx')

display(df)
```

```python
#Tratamento rua
df['rua'] = df['rua'].fillna('Não encontrado')

df.rename(columns={'rua':'Localização'}, inplace=True)

display(df)
```

```python
#Tratamento valor
df['valor'] = df['valor'].str.replace('R$ ', '', regex=False).str.replace('A partir de ', '', regex=False).str.replace('A partir de ', '', regex=False).str.replace('.', '', regex=False).str.replace(',', '', regex=False)


df['valor'] = df['valor'].astype(float)

df.rename(columns={'valor':'Valor do Imóvel(R$)'}, inplace=True)

display(df)
```

```python
#Tratamento metragem
df['metragem'] = df['metragem'].str.replace('Tamanho do imóvel', '', regex=False) 

df['metragem'] = df['metragem'].str.replace('m²', '', regex=False) 
df.rename(columns={'metragem':'Metragem(m²)'}, inplace=True)

display(df)
```

```python
#Tratamento quartos
df['quartos'] = df['quartos'].str.replace('Quantidade de quartos', '', regex=False)

df['quartos'] = df['quartos'].replace('Desconhecido', np.nan)

df.rename(columns={'quartos':'Quantidade de Quartos'}, inplace=True)

df['Quantidade de Quartos'] = df['Quantidade de Quartos'].astype(float)

display(df)
```

```python
#Tratamento banheiros
df['banheiros'] = df['banheiros'].str.replace('Quantidade de banheiros', '', regex=False)

df['banheiros'] = df['banheiros'].str.replace('3-4', '3', regex=False)

df['banheiros'] = df['banheiros'].replace('Desconhecido', np.nan)

df.rename(columns={'banheiros':'Quantidade de Banheiros'}, inplace=True)

df['Quantidade de Banheiros'] = df['Quantidade de Banheiros'].astype(float)

display(df)
```

```python
#Tratamento vagas
df['vagas'] = df['vagas'].str.replace('Quantidade de vagas de garagem', '', regex=False)

df['vagas'] = df['vagas'].replace('Desconhecido', 0)


df.rename(columns={'vagas':'Quantidade de Vagas'}, inplace=True)

df['Quantidade de Vagas'] = df['Quantidade de Vagas'].astype(float)

display(df)
```

```python
#Tratamento valor_iptu

#df['valor_iptu'] = df['valor_iptu'].replace('Desconhecido', np.nan)

#df['valor_iptu'].fillna(0)

# Função para extrair valores de condomínio e IPTU
def extrair_valores(texto):
    if pd.isna(texto):
        return pd.Series([0, 0])
    
    # Extrair valor do condomínio
    match_cond = re.search(r'Cond\.\s*(?:a partir de )?R\$[\s]?([\d\.]+)', texto)
    cond = float(match_cond.group(1).replace('.', '')) if match_cond else 0

    # Extrair valor do IPTU
    match_iptu = re.search(r'IPTU\s*R\$[\s]?([\d\.]+)', texto)
    iptu = float(match_iptu.group(1).replace('.', '')) if match_iptu else 0

    return pd.Series([cond, iptu])

# Aplicar a função
df[['Valor Condominio', 'Valor IPTU']] = df['valor_iptu'].apply(extrair_valores)

# Exemplo: mostrar as primeiras linhas
display(df)
```
