```python
import pandas as pd
import numpy as  np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_excel("Computadores.xlsx")
```

```python


df = df.dropna(subset=['Marca']) #Mesmo com tratamento marcas não estão preenchendo valor
```

```python
agrupamento_itens = df.groupby('Item')['Preco'].count()

marca_preco_med = df.groupby('Marca')['Preco'].mean()

marca_qtd = df.groupby('Marca')['Preco'].count()

media_preco_itens = df.groupby('Item')['Preco'].mean()

media_preco_RAM = df.groupby('RAM')['Preco'].mean()

media_preco_armazenamento = df.groupby('Armazenamento')['Preco'].mean()

qtd_item_RAM = df.groupby('RAM')['Item'].count()

qtd_itens = df['Item'].count()

qtd_marca = df['Marca'].count()

print(f'{agrupamento_itens}')
print('*'*50)
print(f'{media_preco_itens}')
print('*'*50)
print(f'{media_preco_RAM}')
print('*'*50)
print(f'{media_preco_armazenamento}')
print('*'*50)
print(f'{media_preco_armazenamento}')
print('*'*50)
print(f'{qtd_item_RAM}')
print('*'*50)
print(f'{qtd_itens}')

```

```python
plt.figure(figsize=(10,5))
agrupamento_itens.plot(kind="bar", color = "green")
plt.title("Quantidade Itens")
plt.xlabel("Item")
plt.ylabel("Quantidade")
plt.xticks(rotation = 45)
plt.grid(axis = "y")
plt.show()
```

Análise
O gráfico nos mostra que há mais disponibilidades de monitores e notebooks, significa que são produtos que saem mais ou há melhor aderencia do público

```python
plt.figure(figsize=(10,5))
media_preco_itens.plot(kind='bar', color = 'red')
plt.title("Média Preço")
plt.xlabel("Item")
plt.ylabel("Preço")
plt.xticks(rotation = 45)
plt.grid(axis = "y")
plt.show()
```

Essa é a média de preço por dispositivo, vemos que os dispositivos da apple se sobressaem nesse quesito e monitores e computadores possuem uma média bem baixa

```python
plt.figure(figsize=(10,5))

media_preco_armazenamento.plot(kind='bar', color = 'red')
media_preco_RAM.plot(kind='bar', color = 'blue')
plt.legend()
plt.title("Impacto nos preços")
plt.xlabel("Memória RAM")
plt.ylabel("Preço")
plt.xticks(rotation = 45)
plt.grid(axis = "y")
plt.show()
```

Análise impacto no preço
Podemos ver a média de preços dos dispositivos por memória RAM(azul) e armazenamento(vermelho), e podemos verificar o que impacta mais nos preços em geral é a memória RAM, mas em dispositivos de 12GB e 8GB o que impacta mais é o próprio armazenamento

```python
print(marca_preco_med)

media_marcas = marca_preco_med.mean()
plt.figure(figsize=(14,5))

marca_preco_med.plot(kind='bar', color = 'blue')
plt.axhline(y=media_marcas, 
            color="red", 
            linestyle=":", 
            linewidth=2, 
            label=f"Média:{media_marcas:.2f}")
plt.legend()
plt.title("Média Preços por Marca")
plt.xlabel("Marcas")
plt.ylabel("Preço")
plt.xticks(rotation = 90)
plt.grid(axis = "y")
plt.show()
```

Aqui temos um gráfico com as médias de valores dos produtos por marca, podemos ver que a apple é a marca com produtos mais caros, onde a linha vermelha marca a média dos valores
